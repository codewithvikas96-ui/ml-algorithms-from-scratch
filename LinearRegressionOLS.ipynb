```python
import numpy as np 
import matplotlib.pyplot as plt 
```

```python
class LinearRegressionOLS: 

    def __init__(self): 
        self.coefficients = None 
        self.intercept = None 

    def fit(self, X,y):
        X = np.array(X)
        y = np.array(y)

        n_samples = X.shape[0]

        X_bias = np.c_[np.ones(n_samples), X] 

        # calculate theta
        beta = np.linalg.inv(
            X_bias.T @ X_bias
        ) @ ( X_bias.T @ y )

        self.intercept = beta[0]
        self.coefficients = beta[1:]

        

    def predict(self, X): 
        X = np.array(X)
        return X @ self.coefficients + self.intercept

        
        
```

```python
X = np.array([[1],
              [2],
              [3],
              [4],
              [5]])

y = np.array([5, 7, 9, 11, 13])
```

```python
model = LinearRegressionOLS()
model.fit(X,y)

prediction = model.predict(X)

print(f"Coefficient: {model.coefficients}")
print(f"Intercept: {model.intercept}")
print(f"Prediction: {prediction}")
```

```python
# Visualize 
plt.scatter(X,y, label='Actual Data', color='purple')
plt.plot(X, prediction, 'r--', label='Prediction')
plt.legend()
plt.show()
```
