```python
import numpy as np
```

```python
class KNNRegressor:

    def __init__(self, k=3):
        self.k = k

    def fit(self, X, y):

        self.X_train = np.array(X)
        self.y_train = np.array(y)

    def _euclidean_distance(self, x1, x2):

        return np.sqrt(np.sum((x1 - x2) ** 2))

    def _predict_single(self, sample):

        # Calculate distances
        distances = [
            self._euclidean_distance(sample, x_train)
            for x_train in self.X_train
        ]

        # Indices of k nearest neighbors
        k_indices = np.argsort(distances)[:self.k]

        # Target values of nearest neighbors
        k_values = self.y_train[k_indices]

        # Average target value
        prediction = np.mean(k_values)

        return prediction

    def predict(self, X):

        X = np.array(X)

        predictions = [
            self._predict_single(sample)
            for sample in X
        ]

        return np.array(predictions)
```

```python
X = np.array([
    [1000, 2],
    [1200, 2],
    [1500, 3],
    [1800, 3],
    [2000, 4]
])

y = np.array([200, 230, 300, 360, 420])
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

X_scaled = scaler.fit_transform(X)
```

```python
knn = KNNRegressor(k=3)

knn.fit(X_scaled, y)
```

```python
new_house = np.array([[1600, 3]])

new_house_scaled = scaler.transform(new_house)

prediction = knn.predict(new_house_scaled)

print(prediction)
```
